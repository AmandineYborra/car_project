### Importation des librairies

```python
import pandas as pd 
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Importation des données & changement des '?' en NaN

```python
cars = pd.read_csv('C:/Users/Dell/Desktop/Projet_car/Data.csv', na_values=['?'], sep=',')
cars.head()
```

### Changement des NaN par fillna() 

```python
cars['normalized-losses']=cars['normalized-losses'].fillna(cars['normalized-losses'].mean())
cars.head()
```

```python
cars['stroke']=cars['stroke'].fillna(cars['stroke'].mean())
cars.head()
```

```python
cars['bore']=cars['bore'].fillna(cars['bore'].mean())
cars.head()
```

```python
cars['horsepower']=cars['horsepower'].fillna(cars['horsepower'].mean())
cars.head()
```

```python
cars['peak-rpm']=cars['peak-rpm'].fillna(cars['peak-rpm'].mean())
cars.head()
```

```python
cars['num-of-doors']=cars['num-of-doors'].fillna('four')
cars.head()
```

### Suppression des rows dont les prix sont des NaN et création d'un nouveau DF

### Régression linéaire simple

#### Dummies pour les colonnes type object

#### Importation de la librairie 

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
X=cars2.iloc[:,:-1]
y=cars2['price']
```

```python
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.5, random_state=110)
```

#### Création d'un instance de la classe LinearRegression

```python
lr=LinearRegression().fit(X_train, y_train)
```

```python
print(lr.intercept_)
```

```python
print(lr.coef_)
```

```python
dfC=pd.DataFrame(lr.coef_,X.columns,columns=['coef'])
dfC
```

```python
predict=lr.predict(X_test)
predict
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predict))
print('MSE:', metrics.mean_squared_error(y_test, predict))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predict)))
```
